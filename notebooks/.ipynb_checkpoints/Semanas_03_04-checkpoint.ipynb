```python
"""
================================================================================
INICIAÇÃO CIENTÍFICA - ANÁLISE DE SINAIS sEMG (Eletromiografia de Superfície)
================================================================================
Aluno: Anderson Rafael da Silva Lesniewski
RA: 1991329
Orientador: Dr. Fabio Luiz Bertotti
Data: Novembro 2025

Base de dados: NinaPro Database 1 - Subject 1
Fonte: https://ninapro.hevs.ch/instructions/DB1.html

Objetivo:
Aplicar filtros digitais (passa-baixa, passa-alta e passa-banda)
em sinais sEMG reais, observar os efeitos temporais e espectrais
e comparar quantitativamente a energia RMS antes e depois da filtragem.
================================================================================
"""

# ------------------------------------------------------------------------------
# Configuração inicial
# ------------------------------------------------------------------------------
import os
import glob
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import loadmat
from scipy.signal import butter, filtfilt
from scipy.fft import fft, fftfreq

# Configurações visuais
plt.rcParams['figure.figsize'] = (12, 4)
plt.rcParams['font.size'] = 10

print("="*80)
print(" ANÁLISE DE SINAIS sEMG REAIS - NINAPRO DB1 (SUBJECT 1) ")
print("="*80)
print()
```

```python
# ------------------------------------------------------------------------------
# Leitura dos arquivos .mat do sujeito 1
# ------------------------------------------------------------------------------
pasta_dados = "data/s1"

# Listar os arquivos .mat disponíveis
arquivos_mat = glob.glob(os.path.join(pasta_dados, "*.mat"))
print("Arquivos encontrados:")
for arq in arquivos_mat:
    print(" -", os.path.basename(arq))

# Carregar o primeiro exercício como exemplo
arquivo_exemplo = arquivos_mat[0]
data = loadmat(arquivo_exemplo)

print("\nChaves disponíveis no arquivo:")
print(list(data.keys()))
```

```python
# ------------------------------------------------------------------------------
# Exploração e visualização inicial do sinal EMG
# ------------------------------------------------------------------------------

fs = 2000  # Hz - confirmado no artigo original do NinaPro DB1
emg = data['emg']
print("Formato do EMG:", emg.shape)

# Escolher um canal para visualização (0–9)
canal = 0
sinal = np.squeeze(emg[:, canal])

# Criar vetor de tempo (segundos)
tempo = np.arange(len(sinal)) / fs

# Visualizar apenas os primeiros 3 segundos
amostras_trecho = int(fs * 3)
plt.plot(tempo[:amostras_trecho], sinal[:amostras_trecho], color='royalblue')
plt.title(f"Sinal EMG bruto - Canal {canal+1} ({os.path.basename(arquivo_exemplo)})")
plt.xlabel("Tempo (s)")
plt.ylabel("Amplitude (mV)")
plt.grid(True)
plt.show()

print(f"Duração total do sinal: {tempo[-1]:.1f} s | fs = {fs} Hz")
```

```python
# ------------------------------------------------------------------------------
# Aplicação de filtros digitais Butterworth
# ------------------------------------------------------------------------------

def filtro_passa_baixa(sinal, fs, fc=150, ordem=4):
    nyq = fs / 2
    b, a = butter(ordem, fc / nyq, btype='low')
    return filtfilt(b, a, sinal)

def filtro_passa_alta(sinal, fs, fc=20, ordem=4):
    nyq = fs / 2
    b, a = butter(ordem, fc / nyq, btype='high')
    return filtfilt(b, a, sinal)

def filtro_passa_banda(sinal, fs, low=20, high=150, ordem=4):
    nyq = fs / 2
    b, a = butter(ordem, [low / nyq, high / nyq], btype='band')
    return filtfilt(b, a, sinal)

# Aplicar filtros
sinal_pb = filtro_passa_baixa(sinal, fs)
sinal_pa = filtro_passa_alta(sinal, fs)
sinal_pbanda = filtro_passa_banda(sinal, fs)

# Plot comparativo
plt.figure(figsize=(12, 8))

plt.subplot(4, 1, 1)
plt.plot(tempo[:amostras_trecho], sinal[:amostras_trecho], color='gray')
plt.title("Sinal EMG bruto (sem filtragem)")
plt.ylabel("Amplitude (mV)"); plt.grid(True)

plt.subplot(4, 1, 2)
plt.plot(tempo[:amostras_trecho], sinal_pb[:amostras_trecho], color='royalblue')
plt.title("Filtro Passa-Baixa (fc = 150 Hz)")
plt.ylabel("Amplitude (mV)"); plt.grid(True)

plt.subplot(4, 1, 3)
plt.plot(tempo[:amostras_trecho], sinal_pa[:amostras_trecho], color='seagreen')
plt.title("Filtro Passa-Alta (fc = 20 Hz)")
plt.ylabel("Amplitude (mV)"); plt.grid(True)

plt.subplot(4, 1, 4)
plt.plot(tempo[:amostras_trecho], sinal_pbanda[:amostras_trecho], color='crimson')
plt.title("Filtro Passa-Banda (20–150 Hz)")
plt.xlabel("Tempo (s)"); plt.ylabel("Amplitude (mV)"); plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# ------------------------------------------------------------------------------
# Análise espectral (FFT) antes e depois dos filtros
# ------------------------------------------------------------------------------

def calcular_fft(sinal, fs):
    N = len(sinal)
    freqs = fftfreq(N, 1/fs)
    fft_vals = np.abs(fft(sinal))[:N//2]
    return freqs[:N//2], fft_vals

freq_bruto, fft_bruto = calcular_fft(sinal, fs)
freq_pb, fft_pb = calcular_fft(sinal_pb, fs)
freq_pa, fft_pa = calcular_fft(sinal_pa, fs)
freq_pbanda, fft_pbanda = calcular_fft(sinal_pbanda, fs)

plt.figure(figsize=(12, 8))

plt.subplot(4, 1, 1)
plt.plot(freq_bruto, fft_bruto, color='gray')
plt.title("Espectro do Sinal Bruto"); plt.ylabel("Magnitude"); plt.xlim(0, fs/2); plt.grid(True)

plt.subplot(4, 1, 2)
plt.plot(freq_pb, fft_pb, color='royalblue')
plt.title("Filtro Passa-Baixa (fc = 150 Hz)"); plt.ylabel("Magnitude"); plt.xlim(0, fs/2); plt.grid(True)

plt.subplot(4, 1, 3)
plt.plot(freq_pa, fft_pa, color='seagreen')
plt.title("Filtro Passa-Alta (fc = 20 Hz)"); plt.ylabel("Magnitude"); plt.xlim(0, fs/2); plt.grid(True)

plt.subplot(4, 1, 4)
plt.plot(freq_pbanda, fft_pbanda, color='crimson')
plt.title("Filtro Passa-Banda (20–150 Hz)")
plt.xlabel("Frequência (Hz)"); plt.ylabel("Magnitude")
plt.xlim(0, fs/2); plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# ------------------------------------------------------------------------------
# Análise quantitativa (energia RMS)
# ------------------------------------------------------------------------------

def calcular_rms(sinal):
    return np.sqrt(np.mean(np.square(sinal)))

rms_bruto = calcular_rms(sinal)
rms_pb = calcular_rms(sinal_pb)
rms_pa = calcular_rms(sinal_pa)
rms_pbanda = calcular_rms(sinal_pbanda)

print("="*80)
print(" COMPARATIVO DE ENERGIA RMS (mV)")
print("="*80)
print(f"Sinal Bruto       : {rms_bruto:.4f} mV")
print(f"Filtro Passa-Baixa: {rms_pb:.4f} mV")
print(f"Filtro Passa-Alta : {rms_pa:.4f} mV")
print(f"Filtro Passa-Banda: {rms_pbanda:.4f} mV")
print("="*80)

valores = [rms_bruto, rms_pb, rms_pa, rms_pbanda]
labels = ["Bruto", "Passa-Baixa", "Passa-Alta", "Passa-Banda"]
cores = ["gray", "royalblue", "seagreen", "crimson"]

plt.figure(figsize=(8,4))
plt.bar(labels, valores, color=cores)
plt.title("Comparativo de Energia RMS dos Sinais EMG")
plt.ylabel("RMS (mV)")
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.show()
```
